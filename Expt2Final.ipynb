```python
import numpy as np
import matplotlib.pyplot as plt

# Gradient Descent for Linear Regression
def gradient_descent(x, y):
    m = 0  # slope
    b = 0  # intercept
    n = len(x)  # number of data points

    learning_rate=0.01
    iterations=1000

    # Perform Gradient Descent
    for i in range(iterations):
        # Calculate predictions
        y_pred = m * x + b

        # Calculate the gradients
        dm = (-2/n) * np.sum(x * (y - y_pred))  # gradient for m
        db = (-2/n) * np.sum(y - y_pred)         # gradient for b

        # Update m and b
        m -= learning_rate * dm
        b -= learning_rate * db

        # Print iteration information
        if i % 100 == 0:  # Print every 100 iterations
            MSE = np.mean((y - y_pred)**2)  # Calculate Mean Squared Error (MSE) as the cost
            print(f"Iteration {i}: m = {m:.4f}, b = {b:.4f}, MSE = {MSE:.4f}")

    return m, b

# Generate some example data (you can replace this with real data)
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# Run Gradient Descent to find the best-fit line
m, b = gradient_descent(x, y)

# Print the results
print(f"Best-fit line: y = {m:.2f}x + {b:.2f}")

# Plot the results
plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x, m*x + b, color='red', label=f'y = {m:.2f}x + {b:.2f}')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()






```
